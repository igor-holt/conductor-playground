# Experiment: Retraining Eval Ablation

Objective:
- Compare baseline and candidate cycle outcomes.
- Quantify the grader deltas that matter for phase-1 promotion safety.


```python
from __future__ import annotations

import statistics

baseline = [0.93, 0.91, 0.92]
candidate = [0.79, 0.72, 0.73]
delta = statistics.mean(candidate) - statistics.mean(baseline)
summary = {
    'baseline_avg': round(statistics.mean(baseline), 3),
    'candidate_avg': round(statistics.mean(candidate), 3),
    'mean_delta': round(delta, 3),
}
summary
```

## Vital Signals

- If `mean_delta` falls below `-0.10`, the watchdog window should be reviewed.
- Treat `chemical_name` and `llm_judge` as the load-bearing graders.


```python
grader_scores = {
    'chemical_name': [0.74, 0.75, 0.77],
    'llm_judge': [0.69, 0.71, 0.70],
}
risk_flags = {
    grader: round(statistics.mean(scores), 3) < 0.85
    for grader, scores in grader_scores.items()
}
risk_flags
```

The seeded data intentionally lands in rollback territory so the notebook stays aligned with the worker tests.
